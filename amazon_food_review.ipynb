```python
# $ o u l $ h i f t e r

import pandas as pd
import numpy as np
import seaborn as sns
from matplotlib import pyplot as plt
import warnings
warnings.filterwarnings("ignore")
from IPython.core.interactiveshell import InteractiveShell
InteractiveShell.ast_node_interactivity = "all"
```

```python
# load dataset

db = pd.read_csv("/home/vic/Kaggle/amazon_food_review/Reviews.csv")
```

```python
# check basic insights of data

db.head()
db.describe()
db.columns
db.dtypes
```

```python
# take only useful columns from dataset

db.drop(["ProductId", "UserId", "ProfileName", "HelpfulnessNumerator", "HelpfulnessNumerator", "HelpfulnessDenominator", "Time"],axis=1,inplace=True)
db.head()
```

```python
# shuffle dataset creating train set, validation set and test set
from sklearn.utils import shuffle
db = db.sample(frac=1).reset_index(drop=True)
db = db[:1000]
train = db.sample(frac=0.8, random_state=1).reset_index(drop=True)
test = db.loc[~db.index.isin(train.index)]
```

```python
# check size of train, test and validation and check if all labels exist
len(train)
len(test)
train["Score"].unique()
test["Score"].unique()
```

```python
# making target variable as score and dropping from train set
train_target = train.Score
test_target = test.Score
train.drop(["Score"], axis=1, inplace=True)
test.drop(["Score"], axis=1, inplace=True)
train.head()
test.head()
```

```python
# preprocess text data
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
from nltk.tokenize import RegexpTokenizer
import re
from gensim.models.doc2vec import LabeledSentence
tokenizer = RegexpTokenizer(r'\w+')
STOP_WORDS = stopwords.words("english")
all_text_train = train["Text"]
all_text_test = test["Text"]
stemmer = PorterStemmer()


def cleantext(text):
    tagged_doc = []
    for index, i in enumerate(text):
        text_lower = i.lower()
        tokens = tokenizer.tokenize(text_lower)
        text_stop = [word for word in tokens if word not in STOP_WORDS]
        text_num_free = [re.sub(r'[\d]','',i) for i in text_stop]
        text_stemmed = [stemmer.stem(word) for word in text_num_free]
        text_len = [word for word in text_stemmed if len(word)>1]
        text_tag = LabeledSentence(words=text_len, tags=[index])
        tagged_doc.append(text_tag)
    return tagged_doc
```

```python
# clean training set text
docs_train = cleantext(all_text_train)
docs_test = cleantext(all_text_test)
```

```python
#check document
docs_train[0]
```

```python
# train model doc2vec-dm
from gensim.models.doc2vec import Doc2Vec
import multiprocessing
cores = multiprocessing.cpu_count()


def train_doc2vec_model(tagged_docs):
    model = Doc2Vec(dm=1, size=300, window=10, min_count=2, workers=cores, alpha=0.025, min_alpha=0.025)
    model.build_vocab(tagged_docs)
    model.train(tagged_docs, epochs=10, total_examples=model.corpus_count)
    return model
```

```python
# training
model_train = train_doc2vec_model(docs_train)
model_test = train_doc2vec_model(docs_test)
```

```python
# checking embedded vector
len(model_train.docvecs[0])
len(model_test.docvecs[0])
```

```python
# need to vectorize comments for input to classifier
def vectorize_comments(df, model):
    comments = []
    for i in range(df.shape[0]):
        comments.append(model.docvecs[i])
    return comments
```

```python
# creating a dataframe with vectorized comments
input_form_train = vectorize_comments(train, model_train)
input_form_train = np.asarray(input_form_train)
input_form_test = vectorize_comments(test, model_test)
input_form_test = np.asarray(input_form_test)
```

```python
# feature extracted from PV-DM, training a classifier for classification
from sklearn.model_selection import GridSearchCV, StratifiedKFold
from sklearn.ensemble import RandomForestClassifier
def train_classifier(X,y):
    forest = RandomForestClassifier()
    n_estimators = [250, 300]
    min_samples_split = [2, 5]
    min_samples_leaf = [1, 2, 5]
    max_features = ['sqrt', 'log2']
    n_jobs = [-1]
    parameter_grid = {'n_estimators': n_estimators, 'min_samples_leaf': min_samples_leaf, 'min_samples_split': min_samples_split,
                      'n_jobs': n_jobs, 'max_features': max_features}
    cv = StratifiedKFold(n_splits=5)
    clf = GridSearchCV(forest, param_grid = parameter_grid, scoring='accuracy', cv = cv)
    clf.fit(X,y)
    return clf
```

```python
# getting the classifier
clf = train_classifier(input_form_train, train_target)
```

```python
# evaluating results
clf.best_score_
clf.best_estimator_
clf.best_params_
```

```python
# checking accuracy score on test set
from sklearn.metrics import accuracy_score
pred = clf.predict(input_form_test)
accuracy_score(test_target, pred)
```

```python
# checking using xgboost
import xgboost as xgb
def train_classifier_xgb(X,y):
    xgbclf = xgb.XGBClassifier()
    learning_rate = [0.01]
    n_estimators = [250, 300]
    gamma = [0.05]
    max_depth = [3, 5]
    min_child_weight = [1, 3]
    subsample = [0.6]
    reg_alpha = [0, 0.1]
    n_jobs = [-1]
    cv = StratifiedKFold(n_splits=5)
    parameter_grid = {'learning_rate': learning_rate, 'n_estimators': n_estimators, 'gamma': gamma, 'max_depth': max_depth, 'min_child_weight': min_child_weight, 'subsample': subsample,
                      'reg_alpha': reg_alpha, 'n_jobs': n_jobs}
    clf = GridSearchCV(xgbclf, param_grid=parameter_grid, scoring='accuracy', cv=cv)
    clf.fit(X,y)
    return clf
```

```python
# getting xgb classifier
xgb_clf = train_classifier_xgb(input_form_train, train_target)
```

```python
# checking stats
xgb_clf.best_score_
xgb_clf.best_estimator_
```

```python
# predicting results from xgb
pred_xgb = xgb_clf.predict(input_form_test)
accuracy_score(test_target, pred_xgb)
```
